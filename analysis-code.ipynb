```python
## Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value? 
```

```python
sandwich = df[df['Item'].str.contains('Sandwich')]
crispy = sandwich[sandwich['Item'].str.contains('Crispy Chicken')]
```

```python
crispy
```

```python
grilled1=sandwich[sandwich['Item'].str.contains('Grilled Chicken')]
grilled = grilled1[sandwich['Item'].str.contains('Sandwich')]
grilled
```

```python
avg_grilled_cal = round(grilled.Calories.mean(axis=0), 2)
avg_grilled_cal
```

```python
df.head()
```

```python
list(df)
```

```python
avg_grilled_fat = round(grilled['Total Fat (% Daily Value)'].mean(axis=0), 2)
avg_grilled_sat = round(grilled['Saturated Fat (% Daily Value)'].mean(axis=0), 2)
avg_grilled_chol = round(grilled['Cholesterol (% Daily Value)'].mean(axis=0), 2)
avg_grilled_sod = round(grilled['Sodium (% Daily Value)'].mean(axis=0), 2)
avg_grilled_car = round(grilled['Carbohydrates (% Daily Value)'].mean(axis=0), 2) 

avg_grilled_fib = round(grilled['Dietary Fiber (% Daily Value)'].mean(axis=0), 2) 
avg_grilled_A = round(grilled['Vitamin A (% Daily Value)'].mean(axis=0), 2)
avg_grilled_C = round(grilled['Vitamin C (% Daily Value)'].mean(axis=0), 2)
avg_grilled_cal = round(grilled['Calcium (% Daily Value)'].mean(axis=0), 2)
avg_grilled_iron = round(grilled['Iron (% Daily Value)'].mean(axis=0), 2)
```

```python
avg_crispy_fat = round(crispy['Total Fat (% Daily Value)'].mean(axis=0), 2)
avg_crispy_sat = round(crispy['Saturated Fat (% Daily Value)'].mean(axis=0), 2)
avg_crispy_chol = round(crispy['Cholesterol (% Daily Value)'].mean(axis=0), 2)
avg_crispy_sod = round(crispy['Sodium (% Daily Value)'].mean(axis=0), 2)
avg_crispy_car = round(crispy['Carbohydrates (% Daily Value)'].mean(axis=0), 2) 

avg_crispy_fib = round(grilled['Dietary Fiber (% Daily Value)'].mean(axis=0), 2) 
avg_crispy_A = round(crispy['Vitamin A (% Daily Value)'].mean(axis=0), 2)
avg_crispy_C = round(crispy['Vitamin C (% Daily Value)'].mean(axis=0), 2)
avg_crispy_cal = round(crispy['Calcium (% Daily Value)'].mean(axis=0), 2)
avg_crispy_iron = round(crispy['Iron (% Daily Value)'].mean(axis=0), 2)
```

```python
fat = [avg_grilled_fat,avg_crispy_fat]
sat = [avg_grilled_sat,avg_crispy_sat]
chol=[avg_grilled_chol,avg_crispy_chol]
sod=[avg_grilled_sod,avg_crispy_sod]
Car=[avg_grilled_car,avg_crispy_car]
index = ['Grilled','Crispy']
df21 = pd.DataFrame({'Total Fat': fat,
                   'Saturated Fat': sat,
                   'Cholesterol':chol,
                   'Sodium':sod,
                   'Carbohydrates':Car}, index=index)

ax = df21.plot.bar(rot=0,figsize=(10,5))
ax.set_title('Nutrition Value Comparison: Grilled and Crispy Chicken in Sandwich')

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
df21
```

```python
fib=[avg_grilled_fib,avg_crispy_fib]
VitA = [avg_grilled_A, avg_crispy_A]
VitC = [avg_grilled_C, avg_crispy_C]
Cal=[avg_grilled_cal,avg_crispy_cal]
Iron=[avg_grilled_iron,avg_crispy_iron]
index = ['Grilled','Crispy']
df2 = pd.DataFrame({'Fiber':fib,
                    'Vitamin A': VitA,
                   'Vitamin C': VitC,
                   'Calcium':Cal,
                   'Iron':Iron}, index=index)

ax = df2.plot.bar(rot=0,figsize=(10,5))
ax.set_title('Nutrition Value Comparison: Grilled and Crispy Chicken in Sandwich')

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
